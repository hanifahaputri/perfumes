```python
import pandas as pd

df_top = pd.read_csv('Perfumes.csv', usecols = ['Top'])
df_top["Notes"] = df_top["Top"].str.split(",")
df_top = df_top.drop(columns=['Top'])


pd.set_option('display.max_colwidth', 3000)

df_top = df_top.apply( pd.Series.explode )
df_top = df_top.reset_index(drop=True)
df_top.to_csv('Perfumes_Top.csv', sep=',', index=False, encoding='utf-8')
df_top
```

```python
import pandas as pd

df_middle = pd.read_csv('Perfumes.csv', usecols = ['Middle'])
df_middle["Notes"] = df_middle["Middle"].str.split(",")
df_middle = df_middle.drop(columns=['Middle'])

pd.set_option('display.max_colwidth', 3000)

df_middle = df_middle.apply( pd.Series.explode )
df_middle = df_middle.reset_index(drop=True)
df_middle.to_csv('Perfumes_Middle.csv', sep=',', index=False, encoding='utf-8')
df_middle
```

```python
import pandas as pd

df_base = pd.read_csv('Perfumes.csv', usecols = ['Base'])
df_base["Notes"] = df_base["Base"].str.split(",")
df_base = df_base.drop(columns=['Base'])

pd.set_option('display.max_colwidth', 3000)

df_base = df_base.apply( pd.Series.explode )
df_base = df_base.reset_index(drop=True)
df_base.to_csv('Perfumes_Base.csv', sep=',', index=False, encoding='utf-8')
df_base
```

```python
df_all = pd.concat([df_base, df_top, df_middle])
df_all.to_csv('Perfumes_AllNotes.csv', sep=',', index=False, encoding='utf-8')

df_all
```
